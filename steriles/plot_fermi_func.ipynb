```python
import numpy as np
import matplotlib.pyplot as plt
import usphere_utils as uu
```

```python
A, Z = 3, 1
Q = 18
e = np.linspace(0.1,Q,1000)

plt.figure()
plt.plot(e, uu.fermi_func(A, -(Z+1), e))
plt.plot(e, uu.fermi_func(A, (Z+1), e))

plt.figure()
b1 = uu.simple_beta(e, Q, 0)
b1 /= np.max(b1)
b2 = uu.simple_beta(e, Q, 0)*uu.fermi_func(A, -(Z+1), e)
b2 /= np.max(b2)
b3 = uu.simple_beta(e, Q, 0)*uu.fermi_func(A, (Z+1), e)
b3 /= np.max(b3)
plt.plot(e, b1)
plt.plot(e, b2)
plt.plot(e, b3)
plt.show()
```
