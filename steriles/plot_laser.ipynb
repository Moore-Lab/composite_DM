```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d
```

```python
z = np.linspace(-3,10,1000)
w0 = 1e-3 
zR = np.pi* w0**2/1e-3

r = np.linspace(-2,2,2000)

rr,zz = np.meshgrid(r,z)

w = w0*np.sqrt(1 + (zz/zR)**2)

I = (w0/w)**2 * np.exp(-2*rr**2/w**2)
Imax = 1e-3

plt.figure()
for i in range(10):
    print(i)
    npts = int(1e8)
    rand_r = np.random.rand(npts)*4 - 2
    rand_z = np.random.rand(npts)*13 - 3
    w = w0*np.sqrt(1 + (rand_z/zR)**2)
    fvals = (w0/w)**2 * np.exp(-2*rand_r**2/w**2)/Imax
    gpts = np.random.rand(npts) < fvals

    #Iv = I #np.log10(I)
    #Iv[Iv < 1e-8] = np.nan

    #plt.pcolormesh(zz,rr,Iv,cmap='bwr', vmin=-10, vmax=-6,alpha=0.4)
    #plt.colorbar()
    #plt.scatter(zz,rr,c=Iv, rasterized=True, cmap="bwr", vmin=-1e-6, vmax=1e-6)
    plt.scatter(rand_z[gpts], rand_r[gpts], c='r', s=1, alpha=0.01)
plt.axis('off')
plt.savefig('laser.png', transparent=True)
plt.show()
```

```python
z = np.linspace(-0.03,0.03,1000)
w0 = 1.5e-3 
zR = np.pi* w0**2/1e-3

r = np.linspace(-0.06,0.06,2000)

rr,zz = np.meshgrid(r,z)

w = w0*np.sqrt(1 + (zz/zR)**2)

Itot=0
rlist = np.linspace(-0.05,0.05,7)
for roff in rlist:
    I = (w0/w)**2 * np.exp(-2*(rr-roff)**2/w**2)
    I /= np.max(I)
    Itot += I

Iv = np.log10(Itot)
Iv[Iv< -5] = np.nan

plt.figure()
plt.pcolormesh(zz,rr,Iv,cmap='bwr', vmin=-10, vmax=0) #, vmin=-0.01, vmax=0.01,alpha=1)
#plt.axes('equal')
plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('laser_zoom.png', transparent=True)
plt.show()
```
